## LCG(Linear congruential generator)

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from pylab import genfromtxt
import numpy as np


def LCG():
    a = 23
    c = 0
    m = 10**8 + 1
    global rand
    rand = (a*rand + c) % m
    return rand/m


def seed(X_0):
    global rand
    rand = X_0
    
    
N = 1000
Numbers = [i for i in range(N)]
X_0 = 1.5
seed(X_0)

file = f"LCG1-N{N}U{X_0}.txt"



with open(file, 'w') as f:
    for i in Numbers:
        r = LCG()
        f.write(str(r) + "\n")

```

## Plots

```python
with open(file, 'r+') as f:
    x = genfromtxt(f)
    plt.scatter(x, Numbers)
    
    
plt.xlabel('N', fontsize = 20)
plt.ylabel(r'$U_N$', fontsize = 20)
plt.show()
```

```python
with open(file, 'r+') as f:
    x = genfromtxt(f)
    x1 = x[1:]
    x2 = x[:-1]

    plt.scatter(x1, x2)
    
    
plt.xlabel(r'$X_{n+1}$',fontsize = 20)
plt.ylabel(r'$X_n$', fontsize = 20)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

with open(file, 'r+') as f:
    x = genfromtxt(f)
    x1 = x[:-2]
    x2 = x[1:-1]
    x3 = x[2:]

    ax.scatter(x3,x2,x1)
    
    
ax.set_xlabel(r'$U_{n+2}$',fontsize = 20)
ax.set_ylabel(r'$U_{n+1}$',fontsize = 20)
ax.set_zlabel(r'$U_n$', fontsize = 20)
plt.show()
```

## 3. (Mersenne Twister Generator)

```python

import random
random.seed(a=1, version=2)
file = f"LCG1-N{N}U{U_0}.txt"

N = 1000
Numbers = [i for i in range(N)]


with open(file, 'w') as f:
    for i in Numbers:
        r = random.random()
        f.write(str(r) + "\n")
```
