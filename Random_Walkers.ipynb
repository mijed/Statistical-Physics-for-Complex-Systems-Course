## Random Walkers

```python
import numpy as np
import random
import matplotlib.pyplot as plt 

# Walkers Kate and John
def Randomwalkers1D(n,d): # n number of steps, d - John starting position
   # Kate position
   x_K = 0
   # John position
   x_J = d
    
   xposition_K = [0] 
   xposition_J = [d]
   crossing_points = np.zeros(n+1)

   #timepoints
   global time
   time = np.arange(n+1)
    
   global cross_time_table
   cross_time = 0
    
   first_cross = True
    
   for i in range (1,n+1):
    

       if first_cross:
            if x_K>=x_J:
                cross_time = i
                first_cross = False
    
        
    
    
       step_K = np.random.uniform(0,1)
        
       if step_K < 0.5: # moving right
           x_K += 1
       elif step_K >= 0.5: # moving left
           x_K += -1
            
       step_J = np.random.uniform(0,1)
    
       if step_J < 0.5:   
           x_J += 1
       elif step_J >= 0.5: 
           x_J += -1
           
 
       xposition_K.append(x_K)
       xposition_J.append(x_J)
   return [xposition_K,xposition_J,cross_time]
```

```python
Randwalk = Randomwalkers1D(10000,100) 
plt.plot(time,Randwalk[0],Randwalk[1]) 
plt.title("1D Random Walks")
plt.xlabel("Timestep")
plt.ylabel("Position")
plt.show()
```

## Average time - First Cross


```python
avg_time = 0
d_range = np.arange(0,101,1)
crossing_avg_time = np.zeros(len(d_range))
idx = 0
for d in d_range:
    first_cross_time = np.zeros(1000)
    for i in range(1000):
        Randwalk = Randomwalkers1D(10000,d)
        if d == 0:
            first_cross_time[i] = 0
        else:
            first_cross_time[i] = Randwalk[2]
        
    
    
    avg_time = sum(first_cross_time)/len(first_cross_time)
    crossing_avg_time[idx] = avg_time
    idx += 1

plt.plot(d_range,crossing_avg_time)
plt.ylabel("Timestep")
plt.xlabel("d")
plt.savefig("AvgTime_FC",dpi=400)
plt.show()
        
        
    
```
